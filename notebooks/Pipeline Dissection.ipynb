```python
import pysam
import pyfasta
import pandas as pd
import numpy as np
from copy import copy
%pylab inline
import matplotlib.pyplot as plt
```

### Parse SAM output

Split aligned and unaligned reads, retain all the statistics we care about. Since a unique read is composed of a (umi, seq) pair, we count the number of (umi,seq) pairs we have in both aligned and unaligned reads.

```python
sam_path = "/Users/averes/Projects/Melton/temp_dropseq/reads_369_299.raw.sam"
sam_path = "/Users/averes/Projects/Melton/temp_dropseq/bc1.trimmed.sam"

samfile = pysam.Samfile(sam_path, "r" )
ref_lengths = copy(samfile.lengths)

aligned = []
unaligned = []
for a in samfile:
    
    read_name = a.qname
    split_name = read_name.split(':')
    if len(split_name) == 2:
        umi = split_name[1] #Adrian format
    else:
        umi = split_name[4] #Allon format
    
    if a.tid == -1:
        unaligned.append({
            'umi': umi,
            'seq': a.seq,
            'read_len': len(a.seq)
        })
    else:
        ref = samfile.getrname(a.tid)
        aligned.append({
            'transcript': ref.split('|')[0],
            'gene': ref.split('|')[1],
            'NM': int(a.opt('NM')), #Useful sanity check, but very slow to parse.
            'distance_from_end': ref_lengths[a.tid] - a.aend,
            'umi': umi,
            'seq': a.seq,
            'read_len': len(a.seq),
            'a': a,
        })
aligned = pd.DataFrame(aligned)
unaligned = pd.DataFrame(unaligned)

print('# of aligned (umi, seq) pairs : %d' % len(aligned.groupby(['umi', 'seq', ])))
print('# of unaligned (umi, seq) pairs : %d' % len(unaligned.groupby(['umi', 'seq', ])))
```

Next, we need to retain 1 alignment for each (umi, seq, gene). We sort by distance from end, then group by (umi, seq, gene), then keep only the first element of each group. The first 20 elements look like:

```python
one_read_per_gene = aligned.sort('distance_from_end')
one_read_per_gene = one_read_per_gene.groupby(['umi', 'seq', 'gene'], as_index=False).first()
one_read_per_gene = one_read_per_gene.join(one_read_per_gene.groupby(['umi', 'seq']).size().to_frame('aligns_from_read'), on=['umi', 'seq'])
```

Let's look at the distribution of the number of genes per (umi,seq).

```python
fig = plt.figure(figsize=(16,4))
ax = fig.add_subplot(121, title='Alignment multiplicity', 
                 xlabel='Number of alignments', ylabel='#(umi,seq)',
                 yscale='log')
value_counts = one_read_per_gene.groupby(['umi', 'seq']).size().value_counts()
ax.bar(value_counts.index, value_counts, log=True, edgecolor='none', facecolor='g')
print("#(umi,seq) with a single alignment: %d" % value_counts[1])
print("#(umi,seq) with a more than 1 alignment: %d" % (sum(value_counts)-value_counts[1]))

ax = fig.add_subplot(122, title='Alignment quality', 
                 xlabel='Number of mismatches', ylabel='#alignments',
                 yscale='log')
nm_counts = one_read_per_gene['NM'].value_counts()
ax.bar(nm_counts.index, nm_counts, log=True, edgecolor='none', facecolor='g')
```

### Filter reads with multiple alignments
First, remove alignments that are more than D bp away.

Then, only keep the read if it has less tham M alignments to its name.

```python
def remove_distant_reads(df, d=525):
    #Size of group is stored in each record
    close_aligns = df['distance_from_end']<d
    if 0 < close_aligns.sum():
        return df[close_aligns]
    else:
        return df
    
d_filt_df = pd.concat([unique, 
                          non_unique.groupby(['umi', 'seq'], as_index=False).apply(remove_distant_reads)])
d_filt_df = d_filt_df.join(d_filt_df.groupby(['umi', 'seq']).size().to_frame('nearby_aligns_from_read'), on=['umi', 'seq'])
m_filt_df = d_filt_df[d_filt_df['nearby_aligns_from_read'] <= 4]
m_filt_df['score'] = 1./m_filt_df['nearby_aligns_from_read']
```

### Use Hitting-set solution to assign counts from each UMI.

```python
from copy import copy
from collections import defaultdict

split_ambiguities = True
umi_counts = defaultdict(float)
ambig_umi_counts = defaultdict(float)
ambig_gene_partners = defaultdict(set)

for umi, u_aligns in m_filt_df.groupby('umi'):
    
    
    g0 = set(u_aligns['gene'])    
    r0 = set()
    gene_read_mapping = dict()
    
    for seq, seq_genes in u_aligns.groupby('seq')['gene']:
        r0.add(seq)
        seq_genes = set(seq_genes)
        for g in g0:
            gene_read_mapping[(g,seq)] = float(g in seq_genes)/float(len(seq_genes))
       
    target_genes = {}
    
    while len(r0) > 0:
        #For each gene in g0, compute how many reads point ot it
        gene_contrib = dict((gi, sum(gene_read_mapping[(gi, r)] for r in r0)) for gi in g0)

        #Maximum value
        max_contrib = max(gene_contrib.values())

        #Gene with max contrib
        max_contrib_genes = filter(lambda g: gene_contrib[g]==max_contrib, gene_contrib.keys())

        #Pick a gene. Which doesn't matter until the last step
        g = max_contrib_genes[0]
        # target_genes.add(g)

        for r in copy(r0): #Take a copy of r0 doesn't change as we iterate through it
            if gene_read_mapping[(g, r)]: #Remove any reads from r0 that contributed to the picked gene.
                r0.remove(r)

        if len(max_contrib_genes) > 1:
            # Update the gene contribs based on the new r0, but on the 'old' g0.
            # That is why we remove g from g0 after this step only
            gene_contrib = dict((gi, sum(gene_read_mapping[(gi, r)] for r in r0)) for gi in g0)
            ambig_partners = filter(lambda g: gene_contrib[g]==0, max_contrib_genes)

            if len(ambig_partners) > 1:
                for g_alt in ambig_partners:
                    target_genes[g_alt] = float(len(ambig_partners))
                    ambig_gene_partners[g_alt].add(frozenset(ambig_partners))
        else:
            target_genes[g] = 1.
        g0.remove(g)
    
    for gene, ambigs in target_genes.items():
        split_between = ambigs if split_ambiguities else 1.
        umi_counts[gene] += 1./split_between
        ambig_umi_counts[gene] += (1./split_between if ambigs>1 else 0)
```
