```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../s6d13.counts.txt', sep='\t', index_col=0)
```

```python
gene_stats = df[['Sum_ambig', 'Sum_counts']]
cell_counts = df[df.columns[3:]]
```

```python
cell_counts = cell_counts.drop(np.NaN, 0)
```

```python
cells_tpt = 1000*cell_counts/(cell_counts.sum())
genes_tpt = cells_tpt.T
```

```python
ax = plt.subplot(111, xscale='log', xlabel='Molecules per cell', ylabel='# cells')
ax.hist(cell_counts.sum(), bins=np.logspace(2.9, 5, 20));
```

```python
ax = plt.subplot(111, xscale='log', xlabel='INS (Transcripts per thousand+1)', ylabel='# cells')
ax.hist(cells_tpt.ix['INS']+1, bins=np.logspace(0, 3, 30));
```

```python
ax = plt.subplot(111, xscale='log', yscale='log',
    xlabel='Mean (TPT)', ylabel='# cells')

means = genes_tpt.mean()
var = genes_tpt.var()

ax.scatter(means, var/means)
```
