```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/averes/Projects/Melton/temp_dropseq/s6d13.counts.txt', sep='\t', index_col=0)
```

```python
gene_stats = df[['Sum_ambig', 'Sum_counts']]
cell_counts = df[df.columns[3:]]
```

```python
cell_counts = cell_counts.drop(np.NaN, 0)
```

```python
cells_tpt = 1000*cell_counts/(cell_counts.sum())
genes_tpt = cells_tpt.T
```

```python
ax = plt.subplot(111, xscale='log', xlabel='Molecules per cell', ylabel='# cells')
ax.hist(cell_counts.sum(), bins=np.logspace(2.9, 5, 20));
```

```python
ax = plt.subplot(111, xscale='log', xlabel='INS (Transcripts per thousand+1)', ylabel='# cells')
ax.hist(cells_tpt.ix['INS']+0.1, bins=np.logspace(0, 3, 30))
ax.plot([cells_tpt.ix['INS'].median(), cells_tpt.ix['INS'].median()], [0, 300], 'k--')
```

```python
ax = plt.subplot(111, xscale='log', yscale='log',
    xlabel='Mean (TPT)', ylabel='CV (std/mean)')

means = genes_tpt.mean()
stds = genes_tpt.std()

ax.scatter(means, stds/means, alpha=0.3)
```

```python
ax = plt.subplot(111, xscale='log', yscale='log',
    xlim=[10**3, 10**5], ylim=[10**2.5, 10**4.5],
    xlabel='Number of counts', ylabel='Number of genes detected')

counts = cell_counts.sum()
det_genes = (cell_counts>0).sum()

ax.scatter(counts, det_genes, alpha=0.3)
```

```python
ax = plt.subplot(111, xscale='log', xlabel='Genes detected in cell (+1)', ylabel='# cells')
ax.hist((cell_counts>0).sum()+1, bins=np.logspace(2.5, 4.5, 30));
```

```python

from sklearn.decomposition import PCA, FastICA

pca_components = 20
pca = PCA(n_components=pca_components, whiten=True).fit(genes_tpt)

result = pd.DataFrame(
    pca.transform(np.log10(genes_tpt)+0.01),
    index=genes_tpt.index,
    columns=['pc%s' % (s+1) for s in range(pca_components)])
```

```python
ax = plt.subplot(111, xscale='linear', yscale='linear',
    xlabel='Ins log(TPT+0.01)', ylabel='CV (std/mean)')

ax.scatter(result['pc1'],result['pc2'], alpha=0.3)
```
